```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('us_data.csv')
df.head ()

y = 'total'
x = 'year'
c=df[x]
d=df[y]
m,b=np.polyfit(c,d,1)
y_1=m*c+b
plt.plot(df[x],y_1)
plt.scatter(x= df[x], y = df[y])
plt.title ('Total Homelessness VS Year')
plt.xlabel('Year')
plt.ylabel('Total Amount of Homeless People')
```
